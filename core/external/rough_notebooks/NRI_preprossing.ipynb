```python
import pandas as pd

# Read the NRI Table Counties CSV file
df = pd.read_csv('NRI_Table_Counties.csv')

# List of risk columns we want to keep
risk_columns = [
    'STATE', 'STATEABBRV', 'COUNTY',
    'AVLN_RISKR', 'CFLD_RISKR', 'CWAV_RISKR', 'DRGT_RISKR',
    'ERQK_RISKR', 'HAIL_RISKR', 'HWAV_RISKR', 'HRCN_RISKR',
    'ISTM_RISKR', 'LNDS_RISKR', 'LTNG_RISKR', 'RFLD_RISKR',
    'SWND_RISKR', 'TRND_RISKR', 'TSUN_RISKR', 'VLCN_RISKR',
    'WFIR_RISKR', 'WNTW_RISKR'
]

# Select only the risk columns
risk_df = df[risk_columns]

# Save to a new CSV file
output_file = 'NRI_Ratings_Counties.csv'
risk_df.to_csv(output_file, index=False)

print(f"Data saved to {output_file}")
print("\nFirst few rows of the data:")
print(risk_df.head())
```
