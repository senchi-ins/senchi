```python
import trimesh
import numpy as np

# Load the GLB file
model_path = "model_home.glb"
scene = trimesh.load(model_path)

# If the file is a scene (which GLB files typically are), we need to get all meshes
if isinstance(scene, trimesh.Scene):
    # Combine all meshes into one
    meshes = list(scene.geometry.values())
    vertices = np.vstack([mesh.vertices for mesh in meshes])
else:
    vertices = scene.vertices

# Get the bounding box dimensions
min_coords = vertices.min(axis=0)
max_coords = vertices.max(axis=0)

print("Model Dimensions:")
print(f"X range: {min_coords[0]:.2f} to {max_coords[0]:.2f}")
print(f"Y range: {min_coords[1]:.2f} to {max_coords[1]:.2f}")
print(f"Z range: {min_coords[2]:.2f} to {max_coords[2]:.2f}")

print("\nTotal dimensions:")
print(f"Width (X): {max_coords[0] - min_coords[0]:.2f}")
print(f"Height (Y): {max_coords[1] - min_coords[1]:.2f}")
print(f"Depth (Z): {max_coords[2] - min_coords[2]:.2f}")
```
