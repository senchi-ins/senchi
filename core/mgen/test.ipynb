```python
# Import necessary libraries
import asyncio
import json
from typing import List
from house_labelling import label_house
from IPython.display import Image, display, JSON

async def test_house_labelling(image_paths: List[str]):
    """
    Test function to analyze multiple house images and display results
    
    Args:
        image_paths (List[str]): List of paths to the image files
    """
    # Display all images we're analyzing
    for path in image_paths:
        display(Image(path))
    
    # Get the analysis results
    results = await label_house(image_paths)
    
    if results:
        # Display the JSON output file in a nice format
        print("=== Analysis Results ===")
        with open('house_analysis_result.json', 'r') as f:
            json_data = json.load(f)
            display(JSON(json_data))
            
        print(f"\nFinal House Score: {results['final_score']}/100")
    else:
        print("Analysis failed")

# Example usage in notebook
image_paths = [
    "/Users/Adam/projects/senchi/core/mgen/images/sv_response.png",
    "/Users/Adam/projects/senchi/core/mgen/images/arial_response.png"
]

# Run the test
await test_house_labelling(image_paths)
```
